# Merge And Concat

## Merge

```python
import pandas as pd 
var1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
var2=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["A","S","D","F","G","H"]})
pd.merge(var1,var2,on="A")
```

```python
import pandas as pd 
var1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
var2=pd.DataFrame({"A":[1,2,3,4,5,6],"C":["A","S","D","F","G","H"]})
pd.merge(var1,var2,on="A")
```

```python
import pandas as pd 
var1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
var2=pd.DataFrame({"A":[1,2,3,4,7,8],"B":["A","S","D","F","G","H"]})
pd.merge(var1,var2,on="A")
```

```python
import pandas as pd 
var1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
var2=pd.DataFrame({"A":[1,2,3,4,7,8],"C":["A","S","D","F","G","H"]})
pd.merge(var1,var2,how="right")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
var2=pd.DataFrame({"A":[1,2,3,4,7,8],"C":["A","S","D","F","G","H"]})
pd.merge(var1,var2,how="left")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
var2=pd.DataFrame({"A":[1,2,3,4,7,8],"C":["A","S","D","F","G","H"]})
pd.merge(var1,var2,how="inner")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
var2=pd.DataFrame({"A":[1,2,3,4,7,8],"C":["A","S","D","F","G","H"]})
pd.merge(var1,var2,how="outer",indicator=True)
```

```python
var1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
var2=pd.DataFrame({"A":[1,2,3,4,7,8],"B":["A","S","D","F","G","H"]})
pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
var1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
var2=pd.DataFrame({"A":[1,2,3,4,7,8],"B":["A","S","D","F","G","H"]})
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=["name","python"])
```

```python
sr1=pd.Series([1,2,3,4,5,6])
sr2=pd.Series([11,22,33,44,55,66])
pd.concat([sr1,sr2])
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
df2=pd.DataFrame({"A":[1,2,3,4,7,8],"B":["A","S","D","F","G","H"]})
pd.concat([df1,df2])
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
df2=pd.DataFrame({"A":[1,2,3,4,7,8],"B":["A","S","D","F","G","H"]})
pd.concat([df1,df2],axis=1)
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
df2=pd.DataFrame({"A":[1,2,3,4,7,8],"C":["A","S","D","F","G","H"]})
pd.concat([df1,df2],axis=1)
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
df2=pd.DataFrame({"A":[1,2,3,4],"B":["A","S","D","F"]})
pd.concat([df1,df2],axis=1)
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
df2=pd.DataFrame({"A":[1,2,3,4],"C":["A","S","D","F"]})
pd.concat([df1,df2],axis=1,join="inner")
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
df2=pd.DataFrame({"A":[1,2,3,4,7,8],"C":["A","S","D","F","G","H"]})
pd.concat([df1,df2],axis=1,keys=["d1","d2"])
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5,6],"B":["a","b","c","d","e","f"]})
df2=pd.DataFrame({"A":[1,2,3,4,7,8],"C":["A","S","D","F","G","H"]})
pd.concat([df1,df2],axis=0,keys=["d1","d2"])
```

```python
df1=pd.DataFrame({"A":[1,2,3,4,5,6]})
df2=pd.DataFrame({"A":[1,2,3,4,7,8],"B":["A","S","D","F","G","H"]})
pd.concat([df1,df2])
```
