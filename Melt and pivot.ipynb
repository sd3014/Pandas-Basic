# Melt and Pivot

## Melt

```python
import pandas as pd
dic={"days":[1,2,3,4,5,6],"Eng":[13,14,15,16,17,18],"Maths":[7,8,9,10,11,12]}
df=pd.DataFrame(dic)
df
```

```python
pd.melt(df,id_vars="days",var_name="Subjects",value_name="Marks")
```

## Pivot

```python
dic1={"days":[1,2,3,4,5,6],"Stu_name":["a","b","b","a","b","a"],"Eng":[13,14,15,16,17,18],"Maths":[7,8,9,10,11,12]}
df1=pd.DataFrame(dic1)
df1
```

```python
df1.pivot(index="days",columns="Stu_name")
```

```python
dic1={"days":[1,2,1,1,1,2],"Stu_name":["a","b","b","a","b","a"],"Eng":[13,14,15,16,17,18],"Maths":[7,8,9,10,11,12]}
df1=pd.DataFrame(dic1)
df1
```

```python
df1.pivot(index="days",columns="Stu_name",values="Eng")
```

```python
df1.pivot_table(index="Stu_name",columns="days",aggfunc="mean")
```

```python
df1.pivot_table(index="Stu_name",columns="days",aggfunc="sum",margins=True)
```
